# Function for Finding if the payment was done within 30 days

```python
def paid_in_30(account_log,ind,df,days = 31):
    # Intialize
    paid_30 = 'No'
    paybackdate = 'NA'
    partial_30 = 'No'
    # Payement window of 30 days
    new_window = list(account_log.iloc[ind:ind+days,:]['Deliquent_days'])
    
    # Finding index where payments were done
    t = pd.Series([new_window[i] - new_window[i+1] for i in range(len(new_window)-1)])
    
    # Checking of the window is equal to the days before the next due date
    ## Ex in some cases the payment was due after 15 days but only 10 days data was available
    if len(t) < (days-1):
        return('NA','NA','NA')

    # Checking if any payment was made in the remaining days window.
    if (t > 0).any():
        for index in list(t[t > 0].index):
            # Collecting the payment information of that account on that date
            payment = account_log.iloc[ind + index + 1,:]
            paybackdate = payment.Monitor_date
            try:
                # Checking if full payment was made if yes then we flag as payment wast done within 30 days
                ## If not then we flag it that partial payment was made and thus it is still deliquent till we encounter a full payment
                ## In the same time frame
                if payments[(payments.masked_account == payment.masked_account) & (payments.PAYMENT_DATE == payment.Monitor_date)].Full_Payment_Flag.values[0] == 1:
                    paid_30 ='Yes'
                    break
                else:
                    partial_30 = 'Yes'
                    
            except:
                paid_30 = 'Payment_Record_Missing'
                continue 
    return(paid_30,paybackdate,partial_30)
```

# Function for Finding if the any payment was done after 30 days

```python
def paid_any(account_log,ind,df,is_last = False):
    paid_back = 'No'
    paybackdate = 'NA'
    partial = 'No'
    any_payment = list(account_log.iloc[ind:,:]['Deliquent_days'])
    t = pd.Series([any_payment[i] - any_payment[i+1] for i in range(len(any_payment)-1)])
    if (t > 0).any():
        for index in list(t[t > 0].index):
            payment = account_log.iloc[ind + index + 1,:]
            paybackdate = payment.Monitor_date
            try:
                if payments[(payments.masked_account == payment.masked_account) & (payments.PAYMENT_DATE == payment.Monitor_date)].Full_Payment_Flag.values[0] == 1:
                    paid_back = 'Yes'
                    break
                else:
                    partial = 'Yes'
                    if is_last == False:
                        break
            except:
                partial = 'Yes'
                continue
    return(paid_back,paybackdate,partial)
```

#  Now collecting the Information for every account which went deliquent and if they paid back in 30 days

# If not then when they paid back or If they didnt pay even after 30 days of deliquency

```python
deliquency_log = []
for account in list(dpd.groupby(by = 'masked_account').count().index):
    temp = dpd[dpd['masked_account'] == account].reset_index()
    observations = []
    became_deli = temp[temp['Deliquent_days'] == 0]
    became_deli.drop('index',axis = 1,inplace = True)
    if list(temp['Deliquent_days'])[0] > 0 :
        became_deli = became_deli.append(pd.Series(temp.iloc[0,:])).sort_values(by = 'index')

    if (temp['Deliquent_days'] > 0).any():
        for i in list(became_deli.index):
            partial_paydate= 'NA'
            if i == list(became_deli.index)[-1]:
                is_last = True
            else:
                is_last = False


            if became_deli['Deliquent_days'][i] >= 30:
                paid_30 = 'No'
                partial_30 = 'No'
                partial_paydate = 'NA'
                paid_back,paybackdate,partial = paid_any(temp,i,became_deli,is_last)
                if partial == 'Yes':
                    partial_paydate = paybackdate
                    paybackdate = 'NA'
            else:
                paid_30,paybackdate,partial_30 = paid_in_30(temp,i,became_deli,days = 31 - became_deli['Deliquent_days'][i])       
                if partial_30 == 'Yes':
                    partial_paydate = paybackdate
                    paybackdate = 'NA'


                if paid_30 == 'No':
                    paid_back,paybackdate,partial = paid_any(temp,i,became_deli,is_last)
                    if partial in ['Yes','NA']:
                        partial_paydate = paybackdate
                        paybackdate = 'NA'
                elif paid_30 =='Yes':
                    paid_back = 'Yes'
                else:
                    partial = 'NA'
                    paid_back = 'NA'


            observations.append([became_deli.masked_account[i],became_deli.Monitor_date[i],became_deli.Deliquent_days[i]
                                 ,became_deli.Ammount_outstanding[i],paid_30,partial_30,
                                 paid_back,partial,partial_paydate,paybackdate])

    df_log = pd.DataFrame(observations,columns = ['masked_account','Monitor_date','Deliquent_days',
                                           'Ammount_outstanding','Paid_In_30','Parital_30',
                                                  'Paid_back','Partial','partial_paydate','Paybackdate'])
    deliquency_log.append(df_log)
```

```python
Deliquency_log = pd.concat(deliquency_log,axis =0)
```

# While parsing the payments, In some cases we found that some payment information  was missing so we have flagged it as payment missing and handled it in further steps

```python
payment_record_error = set(Deliquency_log[Deliquency_log.Paid_In_30 == 'Payment_Record_Missing'].masked_account)
```

# After collecting log of deliquency of the accounts which went deliquent at some point of time.
# We will now form our data set for ML model
# We will check for each account and if it has failed to pay within 30 days in any of the log of 90 days for which it was observed we will take that reading as negative class
# While those who didn't fail to pay within 30 days of getting deliquent we will take the most oldest reading of that account as postive class

```python
final_ac_deliquency = []
for account in list(set(Deliquency_log.masked_account)):
    ac_log = Deliquency_log[Deliquency_log.masked_account == account]
    if account in list(payment_record_error):
        if (ac_log.Paid_In_30 == 'No').any():
            ac_log.Paid_In_30 = ac_log.Paid_In_30.replace('Payment_Record_Missing','No')

        elif (ac_log[ac_log.Paid_In_30 == 'Payment_Record_Missing'].Paid_In_30 == 'NA').all():
            ac_log.Paid_In_30 = ac_log.Paid_In_30.replace('Payment_Record_Missing','No')

        else:
            ac_log.Paid_In_30 = ac_log.Paid_In_30.replace('Payment_Record_Missing','Yes')

    if (ac_log.Paid_In_30 == 'No').any():
        ac_log = ac_log[ac_log.Paid_In_30 == 'No'].sort_values(by = 'Monitor_date').reset_index().drop(['index'],axis = 1)

    else:
        ac_log = ac_log.sort_values(by = 'Monitor_date').reset_index().drop(['index'],axis = 1)

    ac_log = ac_log.iloc[0,:]    

    if ac_log.Paid_In_30 == 'NA':
        undeliquency_rate = sum(temp.Deliquent_days < 0)/len(temp.Deliquent_days)
        if undeliquency_rate > 0.8:
            ac_log.Paid_In_30 = ac_log.Paid_In_30 = 'Yes'
        else:
            ac_log.Paid_In_30 = ac_log.Paid_In_30 = 'No'

    final_ac_deliquency.append(list(ac_log.values))
```

```python
final_ac_log = pd.DataFrame(final_ac_deliquency,columns = ['masked_account','Monitor_date','Deliquent_days',
                                       'Ammount_outstanding','Paid_In_30','Parital_30',
                                              'Paid_back','Partial','partial_paydate','Paybackdate'])

final_ac_log.Paid_In_30.value_counts()
```
